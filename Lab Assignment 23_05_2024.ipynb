 ### Question 1) Create a Basic Feedforward neural network for binary classification

```python
# Import libraries
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# Generate Random sequence
np.random.seed(0)
X = np.random.rand(100, 2)  
y = (X[:, 0] + 2*X[:, 1] > 1).astype(int)  
```

```python
X
```

```python
y
```

```python
## Initialize Neural Network
model1 = Sequential()

# Adding Layers
model1.add(Dense(3, input_dim=2, activation='relu'))  # Hidden layer with 3 neurons
model1.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron (binary classification)
```

```python
# Compiling model
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Model Training
model1.fit(X, y, epochs=100, batch_size=25)
```

```python
# Evaluating model
loss, accuracy = model1.evaluate(X, y)
print(f'Loss: {loss:.4f}')
print(f'Accuracy: {accuracy*100:.2f}%')
```

### Question 4) Use Keras to create and train basic RNN for sequence prediction. Use sequence (8,9,10,11,12,13)

```python
# Import libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
# Generate sequence
def generate_data():
    X = []
    y = []
    for i in range(1000):
        start = np.random.randint(0, 100)
        sequence = np.array([start + i for i in range(6)])
        X.append(sequence)
        y.append(start + 6)
    return np.array(X), np.array(y)

X, y = generate_data()
```

```python
X
```

```python
y
```

```python
# Reshape sequence(X,y) to be [samples, time steps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))
y = y.reshape((y.shape[0], 1))

print(f'X shape: {X.shape}, y shape: {y.shape}')
```

```python
# Initializing RNN
sequence_model = Sequential([
    SimpleRNN(units=50, input_shape=(X.shape[1], X.shape[2]), activation='relu'),
    Dense(1)
])

# Compiling RNN model
sequence_model.compile(optimizer='adam', loss='mean_squared_error')

# Training the RNN model
sequence_model.fit(X, y, epochs=50, batch_size=10)
```

```python
# Model testing
test_data_sequence = np.array([[i+j for j in range(6)] for i in range(8, 18)])
test_data_sequence = test_data_sequence.reshape((test_data_sequence.shape[0], test_data_sequence.shape[1], 1))
output = sequence_model.predict(test_data_sequence)

# Prediction
print("Predicted Output:")
print(output.flatten())
```

### Question 5) Create basic RNN for text generation using tensorflow. Input text is "Anna, embrace change: Change is constant"

```python
# Import libraries
import numpy as np
import tensorflow as tf
```

```python
# Sample text 
text = "Anna, embrace change: Change is constant."

# Creating vocabulary
vocab = sorted(set(text))
vocab
```

```python
# Create mappings from characters to indices and vice versa
char_to_idx = {char: idx for idx, char in enumerate(vocab)}
print(char_to_idx)
```

```python
idx_to_char = np.array(vocab)
print(idx_to_char)
```

```python
# Text to numerical data
text_as_int = np.array([char_to_idx[char] for char in text])
text_as_int
```

```python
# Define the RNN model
embedding_dim = 256
rnn_units = 1024

# Creating function for model
def build_model(vocab_size, embedding_dim, rnn_units, batch_size):
    model = tf.keras.Sequential([
        tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=None),
        tf.keras.layers.SimpleRNN(rnn_units, return_sequences=True, recurrent_initializer='glorot_uniform'),
        tf.keras.layers.Dense(vocab_size)
    ])
    return model

model = build_model(
    vocab_size=len(vocab),
    embedding_dim=embedding_dim,
    rnn_units=rnn_units,
    batch_size=1
)

# Compiling the model
model.compile(optimizer='adam', loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True))
```

```python
# Number of iterations
num_epochs = 100

# Reshape input data to include batches of sequences
input_text = text_as_int[:-1]
target_text = text_as_int[1:]
print("Input text is \n",input_text)
print("target_text is \n",target_text)
```

```python
# Reshaping for RNN
input_text = np.expand_dims(input_text, axis=0)
print(input_text)
target_text = np.expand_dims(target_text, axis=0)
print(target_text)
```

```python
# Training RNN
model.fit(input_text, target_text, epochs=num_epochs)
```

```python
# Generate text
def generate_text(model, start_string, num_generate=500):
    input_eval = [char_to_idx[s] for s in start_string]
    input_eval = tf.expand_dims(input_eval, 0)
    text_generated = []

    #model.reset_states()
    for _ in range(num_generate):
        predictions = model(input_eval)
        predictions = tf.squeeze(predictions, 0)
        predicted_id = tf.random.categorical(predictions, num_samples=1)[-1, 0].numpy()
        input_eval = tf.expand_dims([predicted_id], 0)
        text_generated.append(idx_to_char[predicted_id])

    return start_string + ''.join(text_generated)

# Generate text
generated_text = generate_text(model, start_string="Anna, embrace change: Change is", num_generate=250)
print(generated_text)
```

### Create LSTM for text analysis
##### Input = ("Anna, embrace change","Change is constant.","She is a positive thinker","Anna takes life as a challenge")

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Sample larger input text corpus
corpus = [
    "Anna, embrace change",
    "Change is constant.",
    "She is a positive thinker",
    "Anna takes life as a challenge"
]

# Prepare tokenizer and word index
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
word_index = tokenizer.word_index
total_words = len(word_index) + 1

# Prepare input-output pairs
sequences = []
for line in corpus:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        sequences.append(n_gram_sequence)

# Pad sequences and create predictors and label
max_sequence_len = max([len(seq) for seq in sequences])
sequences = pad_sequences(sequences, maxlen=max_sequence_len, padding='pre')
x_train, y_train = sequences[:,:-1], sequences[:,-1]

# One-hot encode the labels
y_train = tf.keras.utils.to_categorical(y_train, num_classes=total_words)

# Build LSTM model
model = Sequential()
model.add(Embedding(total_words, 50, input_length=max_sequence_len-1))  # Embedding layer with 50 dimensions
model.add(LSTM(100))
model.add(Dense(total_words, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.01))

# Train the model
model.fit(x_train, y_train, epochs=500, verbose=1)

# Function to generate text with temperature
def generate_text(seed_text, next_words, model, tokenizer, max_sequence_len, temperature=1.0):
    output_text = seed_text
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([output_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        predicted_probs = model.predict(token_list, verbose=0)[0]
        predicted_probs = np.asarray(predicted_probs).astype('float64')
        predicted_probs = np.log(predicted_probs) / temperature
        exp_preds = np.exp(predicted_probs)
        predicted_probs = exp_preds / np.sum(exp_preds)
        predicted_word_index = np.random.choice(range(len(predicted_probs)), p=predicted_probs)
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_word_index:
                output_word = word
                break
        output_text += " " + output_word
    return output_text

# Function to generate multiple sentences
def generate_sentences(seed_text, num_sentences, words_per_sentence, model, tokenizer, max_sequence_len, temperature=1.0):
    sentences = []
    for _ in range(num_sentences):
        sentence = generate_text(seed_text, words_per_sentence, model, tokenizer, max_sequence_len, temperature)
        sentences.append(sentence.strip())
    return ', '.join(sentences)

# Generate multiple sentences
seed_text = "Anna"
num_sentences = 5
words_per_sentence = 10
temperature = 1.0  # Adjust temperature for more or less randomness
generated_text = generate_sentences(seed_text, num_sentences, words_per_sentence, model, tokenizer, max_sequence_len, temperature)
print(generated_text)
```
